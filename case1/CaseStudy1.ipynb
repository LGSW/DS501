# Problem 2: Analyzing Tweets and Tweet Entities with Frequency Analysis

**1. Word Count:** 
* Use the tweets you collected in Problem 1, and compute the frequencies of the words being used in these tweets. 
* Plot a table of the top 30 words with their counts

```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
# Import the necessary package to process data in JSON format

from collections import Counter
from prettytable import PrettyTable
import json

# We use the file saved from last step
tweets_filename = 'iphone_data.txt'
tweets_file = open(tweets_filename, "r")
data = tweets_file.read()
iphone_data = json.loads(data)

ip_texts = [ da['text'] 
                 for da in iphone_data]
hashtags = [ hashtag['text'] 
             for da in iphone_data
                 for hashtag in da['entities']['hashtags'] ]

words = [ w 
          for t in ip_texts 
              for w in t.split() ]



label='Word'
pt = PrettyTable(field_names=[label, 'Count']) 
c = Counter(words)
print(c)
[ pt.add_row(kv) for kv in c.most_common()[:30] ]
pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
print(pt)









```

**2. Find the most popular tweets in your collection of tweets**

Please plot a table of the top 10 tweets that are the most popular among your collection, i.e., the tweets with the largest number of retweet counts.


```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary


retweets = [
            # Store out a tuple of these three values ...
            (da['retweeted_status']['retweet_count'], 
             da['retweeted_status']['user']['screen_name'],
             da['text']) 
            # ... for each status ...
            for da in iphone_data 
            
            # ... so long as the status meets this condition.
                if 'retweeted_status' in da
           ]

pt = PrettyTable(field_names=['Count', 'Screen Name', 'Text'])
[ pt.add_row(row) for row in sorted(retweets, reverse=True)[:10] ]
pt.max_width['Text'] = 60
pt.align= 'l'
print(pt)








```

**3. Find the most popular Tweet Entities in your collection of tweets**

Please plot a table of the top 10 hashtags, top 10 user mentions that are the most popular in your collection of tweets.

```python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
hashtags = [ hashtag['text'] 
             for da in iphone_data
                 for hashtag in da['entities']['hashtags'] ]

usermentions = [ mention['name']
                 for da in iphone_data
                    for mention in da['entities']['user_mentions'] ]
               
label='Word'
pt = PrettyTable(field_names=[label, 'Count']) 
c = Counter(words)
print(c)
[ pt.add_row(kv) for kv in c.most_common()[:30] ]
pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
print(pt)
    
    
label='hashtags'    
pt = PrettyTable(field_names=[label, 'Count']) 
c = Counter(hashtags)
[ pt.add_row(kv) for kv in c.most_common()[:10] ]
pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
print(pt)

label2='user mentions'
pt = PrettyTable(field_names=[label2, 'Count']) 
c = Counter(usermentions)
[ pt.add_row(kv) for kv in c.most_common()[:10] ]
pt.max_width['user mentions'] = 30000
pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
print(pt)

```
